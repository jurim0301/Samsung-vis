## 02. Pandas data frame visualization
- pandas dataframe 다큐먼트 (parameters를 보고 다양한 설정 바꿔보기)
    - http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html
- pandas colormaps 에서의 컬러 팔레트를 밑의 plot 부분에 있는 colormap=에 설정하기
    - http://matplotlib.org/examples/color/colormaps_reference.html
- Pandas의 plot을 통한 시각화는 기본적으로 matplotlib를 사용하는 형태
    - 세세한 설정이 matplotlib와 동일하게 이루어짐
    
- data frame 표를 시각화에 이용해보기
    - basic plotting : 한국의 인구수 변화에 대해 plot
    - heatmap : 전체 dataframe에 대해 인구 수에 따른 heatmap 그려보기
    

#### 한국의 인구수 변화 

```python
ts = df.loc['South Korea']
ts.plot()
```

#### 한국, 일본, 앙고라, 타이완의 인구수 변화 그래프
먼저 여러 행을 가져와서 시각화한 결과 살펴보기

```python
mini_df = df.loc[['South Korea', 'Japan','Angola','Taiwan']]
mini_df.plot(figsize=(10,10))
```

```python
mini_df
```

####  .transpose()
transpose 해서 연도 별 인구수의 변화를 살펴보기

```python
mini_df_ = mini_df.transpose()
mini_df_.plot(title='Population comparison',figsize=(8,8), colormap='seismic')
```

```python
# transpose 한 형태 출력해서 확인해보기
mini_df_
```

#### 기본 barplot 

```python
mini_df_.plot.bar(stacked=True, figsize=(10,10), colormap='rainbow') # stacked=True가 있을 때와 없을 때 비교
```

#### Pandas dataframe자체를 이용해서 heatmap 만들기
- pandas dataframe에는 style이라는 dataframe 전용 스타일러
- 다양하게 표를 분석할 수 있도록 해줌
    - 최솟값, 최댓값 하이라이트
    - 음수, 양수 하이라이트
    - seaborn 스타일 적용

```python
import sys
if sys.version_info[0] < 3: 
    from StringIO import StringIO
else:
    from io import StringIO
    
# seaborn 라이브러리는 유니코드 처리가 어려우므로, 유니코드 문자들 제거한 dataframe 생성
# 나라 이름 중 유니코드가 들어가있는 경우가 있어, 그 경우에는 유니코드 제거한 형태로 출력
def clean_text(txt):
    # return only ascii text
    return StringIO(''.join(txt))
    #return StringIO(''.join([r.decode('unicode_escape').encode('ascii','ignore') for r in txt]))

# dataframe with ascii characters
df_nouni = pd.read_csv(clean_text(open('allCountries_1950_2050','r').read()), sep='\t', index_col=0)
df_nouni
```

```python
# axis = 1 : 행에서 가장 큰 값 하이라이트
# axis = 0 : 열에서 가장 큰 값 하이라이트
df_nouni.style.highlight_max(axis=0)
```

```python
df_nouni.style.highlight_min(axis=1)
```

```python
# cmap으로 컬러팔레트 변경 가능
# seaborn의 경우에는 matplotlib 등과 비교해서 더 다양한 팔레트가 존재함
# http://seaborn.pydata.org/tutorial/color_palettes.html

df_nouni.style.background_gradient(cmap='coolwarm', axis=1)
```
