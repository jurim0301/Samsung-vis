# Word2vec
- 가장 많이 쓰이는 python 라이브러리 gensim을 활용하여 학습시키기
- 학습된 단어들을 시각화

```python
import datetime
import numpy as np

# 시각화 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns

# word2vec 관련(데이터셋 및 모델) 라이브러리
from sklearn import datasets
from gensim.models import Word2Vec

%matplotlib inline
```

# example 1
## 작은 데이터셋으로 학습시켜보기
- 사용자 정의 데이터셋

```python
# data : 한 문장/문서를 단어들의 리스트로 나타낼 때, 그 리스트들의 리스트
# data = [['eat','apple'],['orange','eat'],['eat','rice'],['drink','juice'],['drink','milk'],['drink','water'],\
#        ['orange','juice'],['apple','juice'],['milk','drink'],['water','drink'],['juice','drink'],['orange','eat'],\
#        ['drink','apple','juice'],['eat','orange','apple'],['drink','water','milk'],['drink','orange','juice']]

data = [['eat','apple'],['eat','orange'],['eat','rice'],['drink','juice'],['drink','milk'],['drink','water'],\
       ['orange','juice'],['apple','juice'],['rice','milk'],['milk','drink'],['water','drink'],['juice','drink']]


# size : 벡터의 길이 정의
# min_count : 출현 빈도 제어

# 걸리는 시간 측정
start = datetime.datetime.now()

small_model = Word2Vec(data, size=5, min_count=1)

print('duration:', datetime.datetime.now()-start)
```

```python
small_model.wv.most_similar('apple')
```

```python
small_model.wv['apple']
```

## 작은 데이터셋 단어들 시각화
### Scatter plot

```python
small_model.wv.save_word2vec_format('small.txt')
small = np.loadtxt('small.txt', skiprows=1, dtype='str')
```

```python
vocab = small[:,0]
wordvecs = small[:,1:]
```

```python
# 5차원의 데이터를 2차원으로 나타내는 기법으로 PCA 사용
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(wordvecs)
new_wordvecs = pca.fit_transform(wordvecs)

fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(new_wordvecs[:,0], new_wordvecs[:,1])

for i, txt in enumerate(vocab):
    ax.annotate(txt, (new_wordvecs[i,0], new_wordvecs[i,1]), fontsize=15)

plt.show()
```

### Heatmap

```python
wordvecs
```

```python
plt.plot(figsize=(7,7))
sns.heatmap(wordvecs.astype('float'))
plt.xlabel('dimension')
plt.yticks(np.arange(8)+0.5, vocab, rotation=0, fontsize=15)
plt.show()
```

# example 2
## 큰 데이터셋으로 학습시켜보기
- 20newsgroups 데이터셋

```python
# sklearn 라이브러리에 내재되어있는 20newsgroups 데이터셋 로드
from sklearn.datasets import fetch_20newsgroups

newsgroups = fetch_20newsgroups()['data']

# newsgroups 데이터셋 중 하나 예시로 출력해보기
# 대문자 -> 소문자, 기호 없애기 등의 preprocessing 필요함을 보여줌
print(newsgroups[0])
```

```python
import regex as re

# preprocessing
# regex (regular expression)으로 알파벳, 숫자 이외의 모든 기호들을 제거
# .lower()로 모두 소문자로 변경
# .split()으로 스페이스 단위의 간격으로 문장을 나눔 ex) i saw you -> ['i','saw','you']

data = [re.sub(r'([^\s\w]|_)+','',text).lower().split() for text in newsgroups]
print('total number of documents:', len(data))
```

```python
# 한 문장 출력해보기
print(data[0])
```

```python
start = datetime.datetime.now()
model = Word2Vec(data, size=100, min_count=20)
print('duration:', datetime.datetime.now()-start)
```

```python
len(model.wv.vocab)
```

##  큰 데이터셋 단어들 시각화

```python
model.wv.save_word2vec_format('large.txt')
large = np.loadtxt('large.txt', skiprows=1, dtype='str')
vocab = large[:,0]
wordvecs = large[:,1:]
```

```python
# 100차원의 데이터를 2차원으로 나타내는 기법으로 PCA 사용
pca = PCA(n_components=2)
pca.fit(wordvecs)
new_wordvecs = pca.fit_transform(wordvecs)

fig, ax = plt.subplots(figsize=(8,8))


# 얼마나 출력할 것인지 설정
how_many = 70
ax.scatter(new_wordvecs[:how_many,0], new_wordvecs[:how_many,1])
for i, txt in enumerate(vocab[:how_many]):
    ax.annotate(txt, (new_wordvecs[i,0], new_wordvecs[i,1]), fontsize=15)

plt.show()
```
