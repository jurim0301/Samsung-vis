```python
import scipy.io as sio
from bokeh.plotting import figure, show
from bokeh.layouts import layout, widgetbox
from bokeh.models import ColumnDataSource, HoverTool, Div
from bokeh.models.widgets import Slider, Select, TextInput
from bokeh.io import output_notebook

import time
import numpy as np
```

```python
mat_contents = sio.loadmat('./news.mat')
text_file = open("text.txt", "r")
X = mat_contents['map']
labels = mat_contents['labels']
label = labels[0]
text = text_file.readlines()
```

```python
labels = []
cnt = 0
for i in range(len(label)):
    if label[i] == 0:
        labels.append('business')
    elif label[i] == 1:
        labels.append('Science and technology')
    elif label[i] == 2:
        labels.append('Entertainment')
    elif label[i] == 3:
        labels.append('Health')
    else:
        labels.append('No label')
        cnt += 1
```

```python
nn = 10000
X = X[0:nn, :]
label = label[0:nn]
text = text[0:nn]
labels = labels[0:nn]
```

```python
colormap = {'business': "#ffb400", 'Science and technology': "#0004ff", 'Entertainment': "#fd00cc", 
            'Health':"#f0ff00", 'No label': "#f0ff00"}
colors = [colormap[x] for x in labels]
```

```python
x=X[:, 0]
y=X[:, 1]
radii = np.random.random(size=X.shape[0]) * 1.5
```

```python
TOOLS="crosshair,pan,wheel_zoom,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select"
```

```python
source = ColumnDataSource(
    data = dict(
        x = x,
        y = y,
        labels = labels,
        headlines = text,
        colors = colors,
    )
)
hover = HoverTool(tooltips=[
    ("Category", "@labels"),
    ("Headline", "@headlines")
])
```

```python
p = figure(plot_height=600, plot_width=700, title="NewsAgg dataset by t-SNE", tools=[TOOLS, hover])
```

```python
p.circle(x='x',y='y', source=source, color='colors', radius = 1)
```

```python
output_notebook()
show(p)
```
